```

def pib(c0, c1, g, t0=0, t1=0,  t=0, b0=0, b1=0, i=0):
    
    '''
    c0: consumo autônomo
    c1: propensão marginal a consumir
    g: gastos do governo
    t0: tributo autônomo
    t1: propensão a tributar
    t: tributos
    b0: confiança nos negócios
    b1: propensão a investir
    i: investimento
    '''
    
    y = (1/(1 - c1 + c1*t1 - b1))*(c0 - c1*t0 + b0 + g + i - c1*t)
    M = (1/(1 - c1 + c1*t1 - b1))
    if t == 0:
        T = t0 + t1*y
    else:
        T = t
    
    if i ==0:
        I = b0 + b1*y
    else:
        I = i
    
    print(f'A renda de equilíbrio é igual a: {y:.2f}')
    print(f'Os impostos são iguais a: {T:.2f} ')
    print(f'Os investimentos são: {I:.2f}')
    print(f'A renda disponível é: {(y-T):.2f}')
    print(f'O multiplicador é: {M:.2f}')
    print(f'O consumo é: {(c0+c1*(y-T)):.2f} ')
    
    
    
```
